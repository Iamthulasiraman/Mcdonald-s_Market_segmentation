### **Data Cleaning**

```python
df['Like']= df['Like'].replace({'I hate it!-5': '-5','I love it!+5':'+5'})
```

## **Label encoding : converting yes/no into 1/0**

```python
def labelling(x):
    df[x] = LabelEncoder().fit_transform(df[x])
    return df

comments= ['yummy', 'convenient', 'spicy', 'fattening', 'greasy', 'fast', 'cheap',
       'tasty', 'expensive', 'healthy', 'disgusting']

for i in comments:
    labelling(i)
df
```

## **Considering only first 11 columns**

```python
df_11 = df.loc[:,comments]
df_11
```

```python
x = df_11.values
x
```

## **Principal component analysis**

```python
pca = PCA(n_components=11)
pc = pca.fit_transform(x)
names = ['pc1','pc2','pc3','pc4','pc5','pc6','pc7','pc8','pc9','pc10','pc11']
df_ = pd.DataFrame(data = pc, columns = names)
df_
```

```python
pca.explained_variance_ratio_
```

```python
np.cumsum(pca.explained_variance_ratio_)
```

## **correlation coefficient between original variables and the component**

```python
loadings = pca.components_
num_pc = pca.n_features_
pc_list = ["PC"+str(i) for i in list(range(1, num_pc+1))]
loadings_df = pd.DataFrame.from_dict(dict(zip(pc_list, loadings)))
loadings_df['variable'] = df_11.columns.values
loadings_df = loadings_df.set_index('variable')
loadings_df
```

## **Correlation matrix plot for loadings**

```python
plt.rcParams['figure.figsize'] = (15,10)
ax = sns.heatmap(loadings_df, annot=True)
plt.show()
```

## **Eelbow Method : to find optimal no of clusters**

```python
from bioinfokit.visuz import cluster
cluster.screeplot(obj=[pc_list, pca.explained_variance_ratio_],show=True,dim=(10,5))
```

```python
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer
model = KMeans()
visualizer = KElbowVisualizer(model, k=(1,12)).fit(df_11)
visualizer.show()
```
